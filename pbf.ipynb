```python
import geopandas as gpd
from pyrosm import OSM

# osm = OSM('germany-latest.osm.pbf')
osm = OSM('bremen-latest.osm.pbf')
buildings = osm.get_buildings()



```

```python
# Calculate the area for each building
buildings['area'] = buildings['geometry'].area

# Sort the buildings by area in descending order
sorted_buildings = buildings.sort_values(by='area', ascending=False)

# Print the sorted buildings
print(sorted_buildings)

# Print the biggest building
print("Biggest building:")
print(sorted_buildings.iloc[0])
```

```python
from marian import get_radiation_at_coordinates
import numpy as np
# Load data as 2d array
data = np.loadtxt("grids_germany_annual_radiation_global_2022.asc", skiprows=28)
data[data == -999] = np.nan


def get_radiation_for_building(building):
    # Get the centroid of the building's geometry
    centroid = building['geometry'].centroid

    # Extract latitude and longitude from the centroid
    latitude, longitude = centroid.y, centroid.x

    # Get the radiation value for the building's coordinates
    radiation = get_radiation_at_coordinates(latitude, longitude,data)

    return radiation * building['geometry'].area
```

```python
buildings['radiation'] = buildings.apply(get_radiation_for_building, axis=1)
```
